```python
import pandas as pd
import random
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
df = pd.DataFrame({'whoAmI':lst})
df.head()

```

```python
pd.get_dummies(df['whoAmI'])
```

```python
df.loc[df['whoAmI'] == 'robot', 'robot'] = True
df.loc[df['whoAmI'] != 'robot', 'robot'] = False
df.loc[df['whoAmI'] == 'human', 'human'] = False
df.loc[df['whoAmI'] != 'human', 'human'] = True
df[['robot', 'human']]
# df.head()
```
